```julia
function distance(y_true,y_pred,n)
    result=0
    
    let i = 1 
        while i <= n
            result += abs(y_true[i] - y_pred[i])
            i += 1  # Shorthand for in place increment: x = x + 1
        end
    end
    result
end

    
```

```julia
distance([1,2,3],[1,2,4],3)
```
